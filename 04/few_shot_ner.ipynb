```
import json
from deeppavlov import configs, build_model, train_model

with configs.ner.ner_ontonotes_bert_mult.open(encoding='utf8') as f:
    ner_config = json.load(f)

ner_config['dataset_reader']['data_path'] = './'  # directory with train.txt, valid.txt and test.txt files
ner_config['metadata']['variables']['NER_PATH'] = './'
ner_config['metadata']['download'] = [ner_config['metadata']['download'][-1]]  # do not download the pretrained ontonotes model

ner_model = train_model(ner_config, download=True)
```

```
import pandas as pd
data = pd.read_csv('pristavki.csv', header=None, names=['text'])
```

```
marked = []

for text in data.text.values[:1000]:
    # BERT имеет лимит на длину текста в 512 слов, возьмем даже еще меньше
    if len(text.split()) > 100:
        continue
    pred = ner_model([text])
    sent, tags = pred[0][0], pred[1][0]
    
    # достанем только тексты с сущностями
    if len(set(tags[0])) > 1:
        marked.append(list(zip(sent, tags)))
```

В общем и целом модель научилась весьма сносно. Ниже некоторые примеры её ошибок.

```
# Модель плоховато справляется с "длинным хвостом" в названиях игр. Кроме того, в этом конкретном примере может мешать переключение между языками.

ner_model(['Battlefield 3, Звезды Playstation Битва сильнейших'])
```

```
# Есть странные FP-случаи, которые объясняются недостаточным количество отрицательных данных в датасете.

ner_model(['Kaspersky Anti-Virus – это решение для базовой защиты компьютера от основых видов интернет-угроз.'])
```

```
# Частый "жанр" текстов в данных - перечисление игр с ценами. Между игрой и ценой в данных всегда стоит дефис.
# Если его убрать, получаем FP - "цена" включается в название игры.
# Это даже хорошо, потому как в нормальном случае число после названия франшизы входит в название игры.

ner_model(['Doom 1500, Call of Duty 1200, Tomb Raider 900.'])
```

```
# Иногда название приставки тоже размечается как название игры.

ner_model(['Fifa 08 лицензионный диск x-box 360'])
```

```
# И есть совсем странные необъяснимые ошибки, в которых даже формат нотации IOB2 нарушен.
# В этом примере см. в самом конце непонятно откуда взявшийся тег I-GAME для токена "8гб".

ner_model(['Uncharted Золотая Бездна - 600р/\nNinja Gaiden Sigma Plus - 600р/\nUltimate Marvel vs Capcom 3 - 600р/\nКарта памяти 8гб - 700р'])
```
